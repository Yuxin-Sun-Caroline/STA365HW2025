```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Set sample size
n = 30

# Generate sample data
x = np.random.normal(loc=0, scale=1, size=n)

# Define Bayesian model
with pm.Model() as model1:
    tau = pm.Gamma('tau', alpha=2, beta=1)  # Prior for precision
    theta = pm.Normal('theta', mu=0, sigma=1)  # Prior for mean
    x_obs = pm.Normal('x_obs', mu=theta, tau=tau, observed=x)  # Likelihood
    
    # Sampling from the posterior distribution
    idata1 = pm.sample(1000, chains=2, return_inferencedata=True)

# Plot results
az.plot_posterior(idata1)
plt.tight_layout()
```

### Explanation of Slice Sampling

Slice sampling is a Markov Chain Monte Carlo (MCMC) method that enables us to draw samples from a probability distribution without needing to compute the normalizing constant. The core idea is to introduce an auxiliary variable and sample within a "slice" of the target distribution. The key steps are:

1. **Initialize**: Start with an initial sample value, say $x_1$, within the support of the target probability density function (PDF), $p(x)$.
2. **Sample a horizontal slice**: Draw $y_1$ from a uniform distribution over the range $[0, p(x_1)]$, effectively creating a horizontal slice under the PDF curve.
3. **Find the slice interval**: Identify an interval $[x_L, x_R]$ around $x_1$ where $p(x) > y_1$. This ensures that all candidate values for the next sample fall within the valid density region.
4. **Sample within the slice**: Select a new sample $x_2$ uniformly from the interval $[x_L, x_R]$.
5. **Repeat the process**: Iterate the steps to generate a sequence of samples that approximates the target distribution.

Slice sampling is particularly useful when the full conditional distributions are known only up to a normalizing constant, making it an efficient alternative to Metropolis-Hastings.

---

### Slice Sampling within Metropolis-Within-Gibbs

In a Metropolis-Within-Gibbs framework, we sequentially update each variable while holding others fixed. Slice sampling can replace the Metropolis-Hastings step by ensuring that all proposed values are accepted as long as they satisfy the condition $p(x) > y$.

1. **Initialize**: Choose an initial value $x_1$ from the support of the full conditional distribution.
2. **For each iteration**:
   - Draw an auxiliary variable $y$ uniformly from $[0, p(x_{t-1})]$.
   - Determine an interval $[x_L, x_R]$ such that $p(x) > y$.
   - Sample $x_t$ uniformly from this interval.
3. **Repeat for multiple iterations**, ensuring that each sample is drawn from the correct distribution without rejection.

Unlike Metropolis-Hastings, slice sampling does not require a proposal distribution or rejection step, making it more efficient for high-dimensional problems where full conditionals are difficult to sample from directly.

---

### Understanding the Sampling Curve

The curve being sampled beneath represents the probability density function (PDF) of the target distribution. By defining a horizontal slice at a randomly chosen height $y$, we effectively create a restricted region in which valid samples can be drawn. This ensures that the sampling is proportional to the target distribution while avoiding the need for explicit normalizing constants.

---

### Gibbs Sampling and Dependency

Slice sampling introduces dependency between consecutive samples since each new sample is drawn from a restricted interval determined by the previous sample. This is similar to Metropolis-Hastings, where the Markov chain must mix well to ensure independent draws. To improve efficiency, one can adjust parameters like the slice width $w$ to balance exploration and convergence speed.




This implementation effectively demonstrates slice sampling, where:
- The function samples from a **Beta(1.5,3) distribution**.
- A slice is iteratively defined and sampled from.
- The final histogram approximates the target distribution.

By leveraging slice sampling in a Metropolis-Within-Gibbs setting, we can ensure efficient posterior sampling without rejection, making it particularly valuable for high-dimensional Bayesian inference.


```python
### Python Implementation

#Below is an implementation of slice sampling applied to a Beta distribution:

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def slice_sampling(f, x0, n_samples=1000, w=1.0):
    samples = np.zeros(n_samples)
    x = x0
    for i in range(n_samples):
        y = np.random.uniform(0, f(x))
        x_l, x_r = x - w, x + w
        while f(x_l) > y:
            x_l -= w
        while f(x_r) > y:
            x_r += w
        x = np.random.uniform(x_l, x_r)
        samples[i] = x
    return samples

# Define target distribution (e.g., Beta(1.5,3))
f = lambda x: stats.beta(1.5, 3).pdf(x)
samples = slice_sampling(f, x0=0.5)

# Plot results
plt.hist(samples, bins=30, density=True, label="Slice Sampling")
x_vals = np.linspace(0, 1, 100)
plt.plot(x_vals, f(x_vals), label="Target PDF")
plt.legend()
plt.show()
```

This implementation effectively demonstrates slice sampling, where:
- The function samples from a Beta(1.5,3) distribution.
- A slice is iteratively defined and sampled from.
- The final histogram approximates the target distribution.

By leveraging slice sampling in a Metropolis-Within-Gibbs setting, we can ensure efficient posterior sampling without rejection, making it particularly valuable for high-dimensional Bayesian inference.
